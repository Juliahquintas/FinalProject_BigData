```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("DivisionData").getOrCreate()

df = spark.read.csv(
    "/home/jovyan/work/data/2003.csv.bz2",
    header=True,
    inferSchema=True
)

train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)

print(f"Train rows: {train_df.count()}, Test rows: {test_df.count()}")

# --- Exportamos como CSV ---
train_df.write.mode("overwrite").option("header", True).csv("/home/jovyan/work/data/2003train/")
test_df.write.mode("overwrite").option("header", True).csv("/home/jovyan/work/data/2003test/")
```
