## TRANSFORMACION DE LAS HORAS

```python
from pyspark.sql import functions as F

df = df.withColumn("DepHour", (F.col("CRSDepTime") / 100).cast("int"))
df = df.withColumn(
    "DepHour",
    F.when(F.col("DepHour") == 24, 0).otherwise(F.col("DepHour")))

df = df.withColumn("ArrHour", (F.col("CRSArrTime") / 100).cast("int"))
df = df.withColumn(
    "ArrHour",
    F.when(F.col("ArrHour") == 24, 0).otherwise(F.col("ArrHour")))

df = df.drop("CRSDepTime", "CRSArrTime")
```

```python
# promedio de retraso por hora de salida
dep_delay = (
    df.groupBy("DepHour")
      .agg(F.avg("ArrDelay").alias("avg_delay_dep"))
      .orderBy("DepHour")
      .toPandas()
)

# promedio de retraso por hora de llegada
arr_delay = (
    df.groupBy("ArrHour")
      .agg(F.avg("ArrDelay").alias("avg_delay_arr"))
      .orderBy("ArrHour")
      .toPandas()
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,4))

plt.plot(dep_delay["DepHour"], dep_delay["avg_delay_dep"], label="Salida (DepHour)")
plt.plot(arr_delay["ArrHour"], arr_delay["avg_delay_arr"], label="Llegada (ArrHour)")

plt.xlabel("Hora del día")
plt.ylabel("Retraso promedio (min)")
plt.title("Retraso promedio por hora — salida vs llegada")
plt.legend()
plt.grid(True)

plt.show()
```

```python
from pyspark.sql import functions as F

df = df.withColumn(
    "DepPeriod",
    F.when(F.col("DepHour").between(0,5),  "madrugada")
     .when(F.col("DepHour").between(6,11), "manana")
     .when(F.col("DepHour").between(12,17),"tarde")
     .otherwise("noche")
)

df = df.withColumn(
    "ArrPeriod",
    F.when(F.col("ArrHour").between(0,5),  "madrugada")
     .when(F.col("ArrHour").between(6,11), "manana")
     .when(F.col("ArrHour").between(12,17),"tarde")
     .otherwise("noche")
)

df = df.drop("DepHour", "ArrHour")
```

```python
from pyspark.sql import functions as F
import matplotlib.pyplot as plt

# promedio por franja de salida
dep_period_delay = (
    df.groupBy("DepPeriod")
      .agg(F.avg("ArrDelay").alias("avg_delay"))
      .orderBy("DepPeriod")
      .toPandas()
)

plt.figure(figsize=(7,4))
plt.bar(dep_period_delay["DepPeriod"], dep_period_delay["avg_delay"], color="skyblue")
plt.ylabel("Retraso promedio (min)")
plt.xlabel("Franja horaria de salida")
plt.title("Retraso promedio por franja horaria (DepPeriod)")
plt.show()
```

```python
# promedio por franja de llegada
arr_period_delay = (
    df.groupBy("ArrPeriod")
      .agg(F.avg("ArrDelay").alias("avg_delay"))
      .orderBy("ArrPeriod")
      .toPandas()
)

# gráfico
plt.figure(figsize=(7,4))
plt.bar(arr_period_delay["ArrPeriod"], arr_period_delay["avg_delay"], color="salmon")
plt.ylabel("Retraso promedio (min)")
plt.xlabel("Franja horaria de llegada")
plt.title("Retraso promedio por franja horaria (ArrPeriod)")
plt.show()
```

Puesto que conocemos el retraso de salida del vuelo en minutos, podemos eliminar la columna DepTime, que representa en HH:MM la hora de la salida real del vuelo. Sin embargo, mantenerla no tiene sentido, teniendo la columna de retraso en la salida

```python
df = df.drop("DepTime")
```

```python
print(df.columns)
```

La variable AÑO contiene un único valor, ya que el dataset seleccionado recoge información de un solo año. Por este motivo, esta variable no aporta información relevante y conviene eliminarla.

En cambio, tiene sentido considerar las variables MES, DÍA DEL MES y DÍA DE LA SEMANA, ya que el retraso de un vuelo puede depender de la época del año, del día específico del mes o del día de la semana. Por ejemplo, los retrasos pueden ser más frecuentes durante los fines de semana o en periodos vacacionales.

Veamos la distribución de estas variables para observar posibles patrones.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 fila, 3 columnas

# --- 1. Aeropuertos de salida (Origin) ---
top_origin = (
    df.groupBy("Month")
      .agg(F.avg("ArrDelay").alias("avg_delay"))
      .orderBy(F.desc("avg_delay"))
      .limit(15)
      .toPandas()
)
axes[0].bar(top_origin["Month"], top_origin["avg_delay"], color="skyblue")
axes[0].set_title("MES")
axes[0].set_ylabel("Retraso promedio (min)")
axes[0].tick_params(axis='x', rotation=45)

# --- 2. Aeropuertos de llegada (Dest) ---
top_dest = (
    df.groupBy("DayofMonth")
      .agg(F.avg("ArrDelay").alias("avg_delay"))
      .orderBy(F.desc("avg_delay"))
      .limit(15)
      .toPandas()
)
axes[1].bar(top_dest["DayofMonth"], top_dest["avg_delay"], color="salmon")
axes[1].set_title("DIA DEL MES")
axes[1].tick_params(axis='x', rotation=45)

# --- 3. Aerolíneas (UniqueCarrier) ---
top_carrier = (
    df.groupBy("DayOfWeek")
      .agg(F.avg("ArrDelay").alias("avg_delay"))
      .orderBy(F.desc("avg_delay"))
      .limit(15)
      .toPandas()
)
axes[2].bar(top_carrier["DayOfWeek"], top_carrier["avg_delay"], color="lightgreen")
axes[2].set_title("DIA DE LA SEMANA")
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Se observan diferencias en los retrasos promedios, en base a estas variables, en especial el mes y el dia de la semana. Por ello, se mantendrán y se transforman para que puedan ser utilizadas en el modelo 

```python
df = df.drop("DayOfMonth")
```

### PREPARE DATA FOR MODELING

```python
# Cell 1: Imports
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import matplotlib.pyplot as plt
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.regression import LinearRegression, RandomForestRegressor, GBTRegressor
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
import numpy as np


# Cell 20: Prepare data for modeling
# Index categorical columns
indexers = []
for col_name in ["UniqueCarrier", "Origin", "Dest", "Month", "DayOfWeek", "DepPeriod","ArrPeriod"]:
    indexer = StringIndexer(inputCol=col_name, outputCol=f"{col_name}_index", handleInvalid="keep")
    indexers.append(indexer)

numeric_cols = ["ArrDelay", "DepDelay", "Distance", "TaxiOut"]


# Cell 21: Create pipeline
from pyspark.ml import Pipeline

# Define stages
stages = []

# Add indexers
stages += indexers

# Get indexed column names
indexed_cols = [f"{col}_index" for col in ["ArrDelay", "DepDelay", "Distance", "TaxiOut"]]

# Create vector assembler
assembler = VectorAssembler(
    inputCols=numeric_cols + indexed_cols,
    outputCol="features"
)
stages.append(assembler)

# Create pipeline
pipeline = Pipeline(stages=stages)

# Fit pipeline and transform data
pipeline_model = pipeline.fit(df)
df_transformed = pipeline_model.transform(df)

# Select only needed columns
df_model = df_transformed.select("features", col("ArrDelay").alias("label"))

# Cell 22: Split data
train_data, test_data = df_model.randomSplit([0.8, 0.2], seed=42)
print(f"Training data: {train_data.count()}")
print(f"Test data: {test_data.count()}")

# Cell 23: Train Linear Regression
lr = LinearRegression(featuresCol="features", labelCol="label")

# Train model
lr_model = lr.fit(train_data)

# Make predictions
lr_predictions = lr_model.transform(test_data)

# Evaluate
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")
lr_rmse = evaluator.evaluate(lr_predictions)
print(f"Linear Regression RMSE: {lr_rmse}")

# Cell 24: Train Random Forest
rf = RandomForestRegressor(featuresCol="features", labelCol="label", seed=42)

# Train model
rf_model = rf.fit(train_data)

# Make predictions
rf_predictions = rf_model.transform(test_data)

# Evaluate
rf_rmse = evaluator.evaluate(rf_predictions)
print(f"Random Forest RMSE: {rf_rmse}")

# Cell 25: Train Gradient Boosted Trees
gbt = GBTRegressor(featuresCol="features", labelCol="label", seed=42, maxIter=50)

# Train model
gbt_model = gbt.fit(train_data)

# Make predictions
gbt_predictions = gbt_model.transform(test_data)

# Evaluate
gbt_rmse = evaluator.evaluate(gbt_predictions)
print(f"GBT RMSE: {gbt_rmse}")

# Cell 26: Compare models
print("\n=== Model Comparison ===")
print(f"Linear Regression RMSE: {lr_rmse:.2f}")
print(f"Random Forest RMSE: {rf_rmse:.2f}")
print(f"Gradient Boosted Trees RMSE: {gbt_rmse:.2f}")

# Cell 27: Hyperparameter tuning for best model (Random Forest)
paramGrid = (ParamGridBuilder()
             .addGrid(rf.numTrees, [10, 20, 30])
             .addGrid(rf.maxDepth, [5, 10, 15])
             .build())

crossval = CrossValidator(estimator=rf,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=3,
                          seed=42)

cv_model = crossval.fit(train_data)

# Get best model
best_rf_model = cv_model.bestModel
print(f"Best Random Forest parameters:")
print(f"  Num Trees: {best_rf_model.getNumTrees}")
print(f"  Max Depth: {best_rf_model.getMaxDepth()}")

# Cell 28: Evaluate best model
best_predictions = best_rf_model.transform(test_data)
best_rmse = evaluator.evaluate(best_predictions)
print(f"Best Random Forest RMSE: {best_rmse}")

# Cell 29: Save the best model and pipeline
best_rf_model.write().overwrite().save("best_model_rf")
pipeline_model.write().overwrite().save("preprocessing_pipeline")

print("Model and pipeline saved successfully!")

# Cell 30: Sample predictions
sample_predictions = best_predictions.select("label", "prediction").limit(10).toPandas()
print("\nSample predictions:")
print(sample_predictions)

# Cell 31: Additional metrics
for metric_name in ["rmse", "mae", "r2"]:
    evaluator.setMetricName(metric_name)
    value = evaluator.evaluate(best_predictions)
    print(f"{metric_name.upper()}: {value:.4f}")

# Cell 32: Feature importance
feature_importance = best_rf_model.featureImportances
print("\nTop 10 most important features:")

# Get feature names
feature_names = numeric_cols + indexed_cols
importance_dict = dict(zip(feature_names, feature_importance.toArray()))

# Sort by importance
sorted_importance = sorted(importance_dict.items(), key=lambda x: x[1], reverse=True)
for feature, importance in sorted_importance[:10]:
    print(f"{feature}: {importance:.4f}")

# Cell 33: Visualization of predictions vs actual
comparison_pd = best_predictions.select("label", "prediction").sample(0.01).toPandas()

plt.figure(figsize=(10, 6))
plt.scatter(comparison_pd["label"], comparison_pd["prediction"], alpha=0.5)
plt.plot([-100, 300], [-100, 300], 'r--', label='Perfect prediction')
plt.xlabel("Actual Delay")
plt.ylabel("Predicted Delay")
plt.title("Actual vs Predicted Arrival Delay")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Cell 34: Error distribution
errors = comparison_pd["prediction"] - comparison_pd["label"]
plt.figure(figsize=(10, 6))
plt.hist(errors, bins=50, edgecolor='black')
plt.xlabel("Prediction Error (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Prediction Errors")
plt.grid(True, alpha=0.3)
plt.show()
```
